```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import os
from google.colab import drive
# Change this path to where your file is located
FILE_PATH = '/content/drive/MyDrive/students.txt'
def read_students():
    students = []
    if os.path.exists(FILE_PATH):
        with open(FILE_PATH, 'r') as f:
            for line in f:
                parts = line.strip().split(',')
                if len(parts) >= 5:
                    students.append({
                        'full_name': f"{parts[0]} {parts[1]}",
                        'first': parts[0],
                        'last': parts[1],
                        'email': parts[2],
                        'courses': parts[3],
                        'link': parts[4]
                    })
    return students
students_data = read_students()
student_names = [s['full_name'] for s in students_data]
# --- UI Elements ---
dropdown = widgets.Dropdown(options=student_names, description='Student:')
first_name_w = widgets.Text(description='First Name:', disabled=True)
last_name_w = widgets.Text(description='Last Name:', disabled=True)
email_w = widgets.Text(description='Email:', disabled=True)
courses_w = widgets.Textarea(description='Courses:', disabled=True)
link_w = widgets.Text(description='Link:', disabled=True)
fav_program_w = widgets.Text(description='Fav Program:', placeholder='Enter favorite show...')
update_btn = widgets.Button(description='Update Details', button_style='success')
output = widgets.Output()
# --- Logic ---
def on_student_change(change):
    selected_name = change['new']
    student = next(s for s in students_data if s['full_name'] == selected_name)

    # Update text fields
    first_name_w.value = student['first']
    last_name_w.value = student['last']
    email_w.value = student['email']
    courses_w.value = student['courses']
    link_w.value = student['link']
def on_update_clicked(b):
    with output:
        clear_output()
        new_program = fav_program_w.value
        if not new_program:
            print("Please enter a program name.")
            return

        # Append to the file as requested
        with open(FILE_PATH, 'a') as f:
            f.write(f"\nUpdate for {dropdown.value}: Favorite Program is {new_program}")

        print(f"Successfully updated {dropdown.value}'s favorite program!")
# Register Events
dropdown.observe(on_student_change, names='value')
update_btn.on_click(on_update_clicked)
# --- Display ---
# Grouping for a nicer layout
ui = widgets.VBox([
    widgets.HTML("<h2>Student Information Portal</h2>"),
    dropdown,
    widgets.HBox([first_name_w, last_name_w]),
    email_w,
    courses_w,
    link_w,
    widgets.HTML("<hr>"),
    fav_program_w,
    update_btn,
    output
])
display(ui)
# Trigger initial display
if student_names:
    on_student_change({'new': student_names[0]})
```
